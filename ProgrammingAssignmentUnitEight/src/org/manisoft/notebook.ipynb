```python
import pandas as pd
```

```python
df = pd.read_csv("EmployeeSampleData1.csv")
df
```

```python
df["Annual Salary"] = df["Annual Salary"].str.strip().str.replace("$", "").str.replace(",", "")
df
```

```python
df.dtypes
```

```python
df["Annual Salary"] = pd.to_numeric(df["Annual Salary"])
```

```python
df
```

```python
df["Hire Date"] = pd.to_datetime(df["Hire Date"], format="mixed")
df
```

```python
df.to_csv("EmployeeSampleData1.csv", index=False)
```
